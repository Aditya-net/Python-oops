Numpy Indexing and Slicing of Matrix 

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[3]
```

```python
arr[0,0]
```

```python
arr[0,4]
```

```python
arr[5,4]
```

```python
slice=arr[0:2,1:3]
slice
```

```python
arr[3:6,3:5]
```

```python
arr[3:,3:]
```

```python
arr[:,2]
```

```python
arr[2:4,1:4]
```
