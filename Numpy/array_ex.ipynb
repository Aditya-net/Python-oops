Valid Sudoku

```python
s= np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
s.sum()
```

```python
b=np.sum(s, axis=1)  # Row-wise sum
b == 45  # Check if each row sums to 45
```

```python
b=np.sum(s, axis=1)  # Row-wise sum
for i in b:
    if i!=45:
        print("Not a Sudoku")
        break
else:
    print("Valid Sudoku for Rows")
```

```python
b=np.sum(s, axis=0)  # Column-wise sum
for i in b:
    if i!=45:
        print("Not a Sudoku")
        break
else:
    print("Valid Sudoku for columns")
```

General Questions

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
data.shape  # (5, 3)
```

```python
#Find the average age of students.
avg_age = np.mean(data[:,0])
avg_age  # 18.4
```

```python
#Extract Math marks of all students.
math_marks = data[:,1]
math_marks  # array([85, 92, 76, 65, 90])
```

```python
#Find the highest Science mark.
highest_science_mark = np.max(data[:,2])
highest_science_mark  # 95
```

```python
#Get details of the student who scored more than 90 in Math.
top_math_students = data[data[:,1] > 90]
top_math_students  # array([[19, 92, 88], [20, 90, 85]])
```

```python
#Increase Math marks of all students by 5.
New_math_marks = data[:,1] + 5
New_math_marks  # array([90, 97, 81, 70, 95])
```

```python
#Find how many students are younger than 19.
young_students_count = np.sum(data[:,0]<19)
young_students_count  # 3
```

```python
#Calculate the average marks in each subject (column-wise mean).
average_marks = np.mean(data[:,1:], axis=0)
average_marks  # array([81.6, 83.2])
```

```python
#Get data of students who scored at least 80 in both subjects.
top_students = data[(data[:,1] >= 80) & (data[:,2] >= 80)]
top_students  # array([[19, 92, 88], [20, 90
```
